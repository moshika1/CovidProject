```python
import requests
import csv 
```

```python
response = requests.get('https://coronavirus.m.pipedream.net/')
response 
```

```python
response.json()
```

```python
corona_stats = response.json()

corona_stats
```

```python
corona_stats['rawData']
```

```python
for c in corona_stats['rawData'] :
    print(c['Country_Region'] , '|' , c['Province_State'] , '|' , c['Confirmed'] , '|' , c['Deaths'])
```

```python
with open('corona_stats.csv' , 'w' , newline='') as f :
    writer = csv.DictWriter(f , fieldnames=['Country_Region' , 'Province_State' , 'Confirmed' , 'Deaths'])
    writer.writeheader()

    for c in corona_stats['rawData'] : 
        Country_Region_ = c['Country_Region']
        Province_State = c['Province_State']
        Confirmed = c['Confirmed']
        Deaths = c['Deaths']

        writer.writerow(
            {
                'Country_Region' : Country_Region_ ,
                'Province_State' : Province_State,
                'Confirmed' : Confirmed,
                'Deaths' : Deaths
                 }
                 )


```
